```python
df = df[(df['lon'] >= 116.13) & (df['lon'] < 116.71)]
df = df[(df['lat'] >= 39.77) & (df['lat'] < 40.1)]
```

```python
df.drop(df.loc[df['label'].isin(['motorcycle', 'run', 'boat', 'airplane'])].index, inplace = True)
```

```python
df = df.groupby('user').filter(lambda x: len(x) > 1000)
```

```python
df['year'].value_counts().sort_index()
```

```python
print(df.groupby(['start_go_stop', 'year'])['user'].count())
```

```python
twb_df = df.drop(columns = ['alt', 'time', 'user', 'segment'], axis = 1)
twb_df.head()
```

```python
twb_df.drop(twb_df.loc[twb_df['start_go_stop'].isin(['start', 'stop'])].index, inplace = True)
twb_df['start_go_stop'].value_counts()
```

```python
twb_df.groupby(['year', 'month']).count()
```
